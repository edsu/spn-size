# SPN Sizes

```python
import pandas

spn2 = pandas.read_csv('save-page-now.csv', parse_dates=['date'], index_col='date')
spn2['gb'] = spn2['bytes'] / 1024 / 1024 / 1024
spn2
```

```python
from plotly import express as px

px.line(y=spn2.gb, x=spn2.index, labels={'x': 'Time', 'y': 'GB/day'}, title='SPN2 Daily Ingest')
```

```python
spn1 = pandas.read_csv('liveweb.csv', parse_dates=['date'], index_col='date')
spn1['gb'] = spn1['bytes'] / 1024 / 1024 / 1024
spn1
```

```python
px.line(x=spn1.index, y=spn1.gb, labels={"x": "Time", "y": "GB/day"}, title="SPN1 Ingest Rate")
```

```python
spn = pandas.DataFrame({"spn1": spn1.gb, "spn2": spn2.gb})
spn
```

```python
px.line(spn)
```

```python
spn_month = spn.resample('M').sum()
spn_month
```

```python
fig = px.line(spn_month, title='SPN1 and SPN2 Ingest Rate', labels={'value': 'GB/month', 'date': 'Time'})
fig.update_layout(legend_title='Service Version')
fig
```
