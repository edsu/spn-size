# SPN Sizes

```python
import pandas

spn2 = pandas.read_csv('save-page-now.csv', parse_dates=['date'], index_col='date')
spn2['tb'] = spn2['bytes'] / 1024 / 1024 / 1024 / 1024 
spn2
```

```python
from plotly import express as px
px.line(y=spn2.tb, x=spn2.index, labels={'x': 'Time', 'y': 'TB/day'}, title='SPN2 Daily Ingest')
```

```python
spn1 = pandas.read_csv('liveweb-20211005.csv', parse_dates=['date'], index_col='date')
spn1['tb'] = spn1['bytes'] / 1024 / 1024 / 1024 / 1024
spn1
```

```python
px.line(x=spn1.index, y=spn1.tb, labels={"x": "Time", "y": "TB/day"}, title="SPN1 Ingest Rate")
```

```python
spn = pandas.DataFrame({"liveweb": spn1.tb, "save-page-now": spn2.tb})
spn
```

```python
px.line(spn)
```

```python
spn_month = spn.resample('M').sum()
spn_month
```

```python
fig = px.line(spn_month, title='liveweb and save-page-now collections', labels={'value': 'TB/month', 'date': 'Time'})
fig.update_layout(legend_title='Collection')
```
